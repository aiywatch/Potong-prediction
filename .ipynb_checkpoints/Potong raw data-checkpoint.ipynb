# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime, time
```

```python
dataset = pd.read_csv('data/potong-1.csv.bz2')
data = dataset
data
```

# Dealing with time

```python
timestamp = pd.to_datetime(dataset['timestamp'])
data['timestamp'] = pd.to_datetime(dataset['timestamp'])
```

```python
def get_day_of_week(dt):
    return dt.weekday()
def get_hour(dt):
    return dt.hour
```

```python
day_of_week = timestamp.apply(get_day_of_week)
hour = timestamp.apply(get_hour)

data['day_of_week'] = day_of_week
data['hour'] = hour
```

```python
first_point = (data['trip_id'].shift() != data['trip_id'])
# first_point[0] = False
data['first_point'] = first_point
```

# Calculate time from 1st point

```python
first_points = data[data['first_point']]
started_time_column = []
started_location = []
for index, point in first_points.iterrows():
    length = len(data[data['trip_id'] == point['trip_id']])
    
    started_time_column += [point['timestamp']] * length
    started_location += [point['linear_ref']] * length
```

```python
data['started_time'] = started_time_column
data['started_location'] = started_location
```

```python
data['second_from_started'] = (timestamp - data['started_time']).dt.total_seconds()
```

# Last point

```python
data['last_point_time'] = timestamp.shift()
```

```python
data['second_from_last_point'] = (data['timestamp'] - data['timestamp'].shift()).dt.total_seconds()
```

```python
data['distance_from_last_point'] = data['linear_ref'] - data['linear_ref'].shift()
```

```python
data['last_point_location'] = data['linear_ref'].shift()
```

```python
data
```

# Filter relavant data and divide into in and out trip

```python
data = data[['linear_ref', 'direction', 'day_of_week', 'hour', 'second_from_last_point', 'last_point_location']]
data
```

# Remove Missing data and Outlier

```python
data.dropna(inplace=True)
```

```python
data = data.drop(data[data['distance_from_last_point'] <= 0].index)
```

```python
data = data.drop(data[data['distance_from_last_point'] > 0.2].index)
```

```python
data = data.drop(data[data['second_from_last_point'] > 1000].index)
```

```python
data
```

# Machine Learning - Train data

```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

labelencoder_direction = LabelEncoder()
X[:, 0] = labelencoder_direction.fit_transform(X[:, 0])

onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
```

```python
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc_X = StandardScaler()
# X_train = sc_X.fit_transform(X_train)
# X_test = sc_X.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```
